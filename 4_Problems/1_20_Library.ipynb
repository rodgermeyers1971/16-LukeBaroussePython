**Problem Statement:**
Use the <code>requests</code> library to fetch data from the URL 'https://pokeapi.co/api/v2/pokemon/pikachu' and print the JSON response.

**Hint:**
Use the <code>requests.get()</code> function to fetch data from the URL, providing url as the argument; this will provide a response object.
With the response object, use the <code>json</code> method to get the response formatted as json.

```python
import requests
# Code Needed
url = 'https://pokeapi.co/api/v2/pokemon/pikachu'
dataget = requests.get(url).json()
dataget
```
