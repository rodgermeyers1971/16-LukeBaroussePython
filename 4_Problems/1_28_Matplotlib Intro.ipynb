## ðŸŸ© Simple Line Plot (1.28.1)

**Problem Statement:**
Create a simple line plot with the pyplot module of matplotlib, using the number of job postings over the first 4 months of the year. Use the <code>months</code> as the x-axis and the number of <code>postings</code> as the y-axis. Below are the given variables.

**Hint:**
Use the <code>plt.plot()</code> function to create the line plot.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Code Needed
months = ['January', 'February', 'March', 'April']
postings = [120, 150, 100, 130]

plt.plot(months, postings)
```
