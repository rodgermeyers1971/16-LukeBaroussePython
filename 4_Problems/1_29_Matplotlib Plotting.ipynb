```python
#!pip install datasets
import pandas as pd
from datasets import load_dataset
from matplotlib import pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

## ðŸŸ© Job Country Bar Chart (1.29.1)

**Problem Statement:**
Create a bar chart showing the number of job postings for the <code>job_country</code>. Use the <code>job_country</code> column and count the number of postings for each location. Only get the top 5 job countries. Additionally, change the color of the bars to 'orange'.

**Hint:**
Use the <code>plt.bar()</code> function to create the bar chart.

```python
country_postings = df['job_country'].value_counts().head(5)
plt.bar(country_postings.index, country_postings.values, color = 'orange')
```

## ðŸŸ¨ Average Salary Horizontal Bar Chart (1.29.2)

**Problem Statement:**
Create a horizontal bar chart showing the average yearly salary for different job titles. Use the <code>job_title_short</code> and <code>salary_year_avg</code> columns to calculate the average salary for each job title.

**Hint:**
Use the <code>plt.barh()</code> function to create the horizontal bar chart.

```python
avg_by_job = df.groupby('job_title_short')['salary_year_avg'].mean()
avg_by_job = avg_by_job.sort_values(ascending=True)

plt.barh(avg_by_job.index, avg_by_job.values)
```

## ðŸŸ¥ Average Salary Trend Line Plot (1.29.3)

**Problem Statement:**
Create a line plot showing the trend of average yearly salary over the months of the year. Use the <code>job_posted_date</code> column to extract the month and calculate the average salary for each month.

**Hint:**
Use the <code>plt.plot()</code> function to create the line plot.


```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

df['job_posted_month'] = df['job_posted_date'].dt.month
monthly_avg_salary = df.groupby('job_posted_month')['salary_year_avg'].mean().sort_index()

plt.plot(monthly_avg_salary.index, monthly_avg_salary.values, color='green')
plt.show()
```
